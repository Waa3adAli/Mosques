```python
import random

# Initialize empty board
def create_board():
    return [" " for _ in range(9)]

# Display the board
def display_board(board):
    print()
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print()

# Display position guide
def show_positions():
    print("Position guide (choose from 1 to 9):")
    print("1 | 2 | 3")
    print("--+---+--")
    print("4 | 5 | 6")
    print("--+---+--")
    print("7 | 8 | 9")
    print()

# Check if position is free
def is_free(board, pos):
    return board[pos - 1] == " "

# Check for win
def check_win(board, mark):
    win_positions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]
    for combo in win_positions:
        if all(board[i] == mark for i in combo):
            return True
    return False

# Check for draw
def check_draw(board):
    return " " not in board

# Player move
def player_move(board, mark):
    while True:
        try:
            pos = int(input(f"Enter position for {mark} (1-9): "))
            if 1 <= pos <= 9 and is_free(board, pos):
                board[pos - 1] = mark
                break
            else:
                print("Invalid or taken position.")
        except:
            print("Please enter a valid number between 1 and 9.")

# Computer move (random)
def computer_move(board, mark):
    free_positions = [i + 1 for i in range(9) if board[i] == " "]
    pos = random.choice(free_positions)
    print(f"Computer chooses position {pos}")
    board[pos - 1] = mark

# Main game function
def play_game():
    print("Welcome to Tic-Tac-Toe!")
    mode = input("Play with (1) Friend or (2) Computer? Enter 1 or 2: ")
    show_positions()

    # Player symbol choice
    player1 = input("Choose your symbol (X or O): ").upper()
    player2 = "O" if player1 == "X" else "X"

    turn = input("Who starts first? (1) You or (2) Opponent/Computer: ")

    board = create_board()
    current = "player1" if turn == "1" else "player2"

    while True:
        display_board(board)

        if current == "player1":
            player_move(board, player1)
            if check_win(board, player1):
                display_board(board)
                print("Player 1 wins!")
                break
            current = "player2"
        else:
            if mode == "1":
                player_move(board, player2)
            else:
                computer_move(board, player2)
            if check_win(board, player2):
                display_board(board)
                print("Player 2 wins!" if mode == "1" else "Computer wins!")
                break
            current = "player1"

        if check_draw(board):
            display_board(board)
            print("It's a draw!")
            break

# Play again loop
while True:
    play_game()
    again = input("Play again? (y/n): ").lower()
    if again != 'y':
        print("Goodbye!")
        break
```
